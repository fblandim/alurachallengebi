#### Importing libries to the project

```python
import pandas as pd
import numpy as np
import psycopg2
import os
from io import StringIO
```

#### Importing Datasets

```python
dfIdadeGenero = pd.read_json('Datasets\Tabela_idade_e_genero.json')
```

```python
dfIdadeGenero
```

Replacing (,) to (.)

```python
dfIdadeGenero = dfIdadeGenero.replace({',': '.'}, regex=True)
```

```python
dfIdadeGenero
```

#### Replacing nothing to 0 (this is necessary due to the conversion of columns type to int)

```python
dfIdadeGenero = dfIdadeGenero.replace({'': '0'}, regex=True)
```

```python
dfIdadeGenero
```

#### Spliting column values to the .0 to transform then in intenger numbers, here I am creating an extra column in dataframe tha should be discarted next

```python
dfIdadeGenero[['Cliques no link', 'Cliques no link2']] = dfIdadeGenero['Cliques no link'].str.split('.', expand=True)
dfIdadeGenero[['Visualizações por página', 'Visualizações por página2']] = dfIdadeGenero['Visualizações por página'].str.split('.', expand=True)
dfIdadeGenero[['Compras no website', 'Compras no website2']] = dfIdadeGenero['Compras no website'].str.split('.', expand=True)
dfIdadeGenero[['Adicionados ao carrinho', 'Adicionados ao carrinho2']] = dfIdadeGenero['Adicionados ao carrinho'].str.split('.', expand=True)
dfIdadeGenero[['Checkouts Iniciados', 'Checkouts Iniciados2']] = dfIdadeGenero['Checkouts Iniciados'].str.split('.', expand=True)
dfIdadeGenero[['Compras', 'Compras2']] = dfIdadeGenero['Compras'].str.split('.', expand=True)
```

```python
dfIdadeGenero
```

#### Dropping no valuable columns from dataframe

```python
dfIdadeGenero = dfIdadeGenero.drop(['Cliques no link2','Visualizações por página2','Compras no website2',
                     'Adicionados ao carrinho2','Checkouts Iniciados2','Compras2', 'FIELD4'], axis=1)
```

```python
dfIdadeGenero
```

#### Converting columns type

```python
dfIdadeGenero = dfIdadeGenero.astype({
    
    'Quantia gasta (BRL)' : float,
    'Cliques no link' : int,
    'Visualizações por página' : int,
    'Compras no website' : int, 
    'Compras no facebook' : int,
    'Adicionados ao carrinho' : int,
    'Valor de conversão adicionado ao carrinho' : float,
    'Checkouts Iniciados' : int,
    'Valor de conversão de checkouts iniciados' : int,
    'Valor de conversão de compras' : float,
    'Compras' : int    
})
```

```python
dfIdadeGenero.dtypes
```

#### Replacing NaN in dataframe to 0

```python
dfIdadeGenero = dfIdadeGenero.replace(np.nan, 0)
```

#### Renaming Columns

```python
dfIdadeGenero.columns = ["idade", "genero", "dia", "alcance", 
                         "impressoes", "quantia_gasta_brl", "cliques_no_link", "visualizacoes_por_pagina",
                         "compras_no_website", "compras_no_facebook", "adicionados_ao_carrinho", 
                         "valor_de_conversao_adicionado_ao_carrinho", "checkouts_iniciados",
                         "valor_de_conversao_de_checkouts_iniciados", "valor_de_conversao_de_compras", 
                         "compras"]
```

#### Creating a index name column

```python
index = dfIdadeGenero.index
index.name = "id"
```

#### Connecting to PostgreSQL database

```python
conn = psycopg2.connect(host="localhost", port = 5432, database="db_alurashop", user="postgres", password="123456")
```

#### Creating a dataframe buffer in memory as CSV before insert data into database

```python
sio = StringIO()
sio.write(dfIdadeGenero.to_csv(index=None, header=None)) 
sio.seek(0)
```

#### Inserting data into database using the temporary CSV file

```python
with conn.cursor() as c:
    c.copy_from(sio, "idadegenero", columns=dfIdadeGenero.columns, sep=',')
    conn.commit()
```
