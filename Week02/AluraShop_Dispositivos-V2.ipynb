#### Importing libries to the project

```python
import pandas as pd
import numpy as np
import psycopg2
import os
from io import StringIO
```

#### Importing datasets

```python
dfDispositivos = pd.read_json('Datasets\Tabela_dispositivos.json')
```

```python
dfIdadeGenero = pd.read_json('Datasets\Tabela_idade_e_genero.json')
```

```python
dfDispositivos
```

#### Replacing (,) to (.) 

```python
dfDispositivos = dfDispositivos.replace({',': '.'}, regex=True)
```

```python
dfDispositivos
```

#### Replacing nothing to 0 (this is necessary due to the conversion of columns type to int)

```python
dfDispositivos = dfDispositivos.replace({'': '0'}, regex=True)
```

```python
dfDispositivos
```

#### Spliting column values to the .0 to transform then in intenger numbers, here I am creating an extra column in dataframe tha should be discarted next

```python
dfDispositivos[['Cliques em links', 'Cliques em links2']] = dfDispositivos['Cliques em links'].str.split('.', expand=True)
dfDispositivos[['Visualizações por página', 'Visualizações por página2']] = dfDispositivos['Visualizações por página'].str.split('.', expand=True)
dfDispositivos[['Compras no website', 'Compras no website2']] = dfDispositivos['Compras no website'].str.split('.', expand=True)
dfDispositivos[['Adicionados ao carrinho', 'Adicionados ao carrinho2']] = dfDispositivos['Adicionados ao carrinho'].str.split('.', expand=True)
dfDispositivos[['Checkouts Iniciados', 'Checkouts Iniciados2']] = dfDispositivos['Checkouts Iniciados'].str.split('.', expand=True)
dfDispositivos[['Compras', 'Compras2']] = dfDispositivos['Compras'].str.split('.', expand=True)
```

```python
dfDispositivos
```

#### Dropping no valuable columns from dataframe

```python
dfDispositivos = dfDispositivos.drop(['Cliques em links2','Visualizações por página2','Compras no website2',
                     'Adicionados ao carrinho2','Checkouts Iniciados2','Compras2'], axis=1)
```

#### Converting columns type

```python
dfDispositivos = dfDispositivos.astype({
    
    'Quantia gasta (BRL)' : float,
    'Cliques em links' : int,
    'Visualizações por página' : int,
    'Compras no website' : int, 
    'Compras no facebook' : int,
    'Adicionados ao carrinho' : int,
    'Valor de conversão adicionado ao carrinho' : float,
    'Checkouts Iniciados' : int,
    'Valor de conversão de compras' : float,
    'Compras' : int    
})
```

```python
dfDispositivos.dtypes
```

#### Replacing NaN in dataframe to 0

```python
dfDispositivos = dfDispositivos.replace(np.nan, 0)
```

#### Renaming Columns

```python
dfDispositivos.columns = ["dia", "plataforma_do_dispositivo", "colocacao", "plataforma", "alcance", 
                         "impressoes", "quantia_gasta_brl", "cliques_em_links", "visualizacoes_por_pagina",
                         "compras_no_website", "compras_no_facebook", "adicionados_ao_carrinho", 
                         "valor_de_conversao_adicionado_ao_carrinho", "checkouts_iniciados",
                         "valor_de_conversao_de_checkouts_iniciados", "valor_de_conversao_de_compras", 
                         "compras"]
```

#### Connecting to PostgreSQL database

```python
conn = psycopg2.connect(host="localhost", port = 5432, database="db_alurashop", user="postgres", password="123456")
```

#### Creating a dataframe buffer in memory as CSV before insert data into database

```python
sio = StringIO()
sio.write(dfDispositivos.to_csv(index=None, header=None)) 
sio.seek(0)
```

#### Inserting data into database using the temporary CSV file

```python
with conn.cursor() as c:
    c.copy_from(sio, "dispositivos", columns=dfDispositivos.columns, sep=',')
    conn.commit()
```
